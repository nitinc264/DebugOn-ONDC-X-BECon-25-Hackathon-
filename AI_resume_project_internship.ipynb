```python
import spacy
import pdfplumber
import docx
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load NLP model
nlp = spacy.load("en_core_web_sm")

# Extract text from PDF
def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = " ".join([page.extract_text() for page in pdf.pages if page.extract_text()])
    return text

# Extract text from DOCX
def extract_text_from_docx(docx_path):
    doc = docx.Document(docx_path)
    return " ".join([para.text for para in doc.paragraphs])

# Extract skills using NLP
def extract_skills(text):
    doc = nlp(text)
    return list(set(token.text.lower() for token in doc if token.pos_ in ["NOUN", "PROPN"]))

# Process job description
def process_job_description(job_desc):
    doc = nlp(job_desc)
    return " ".join([token.lemma_ for token in doc if not token.is_stop])

# Train TF-IDF model
def train_model(resumes, job_desc):
    vectorizer = TfidfVectorizer()
    job_vec = vectorizer.fit_transform([job_desc])
    resume_vecs = vectorizer.transform(resumes)
    scores = cosine_similarity(resume_vecs, job_vec).flatten()
    return scores

# Provide job description manually
job_description = "Looking for a Data Scientist with Python, NLP, and Machine Learning skills."

# Read uploaded file
file_path = list(uploaded.keys())[0]  # Get uploaded filename

if file_path.endswith(".pdf"):
    resume_text = extract_text_from_pdf(file_path)
elif file_path.endswith(".docx"):
    resume_text = extract_text_from_docx(file_path)
else:
    raise ValueError("Invalid file format. Please upload a PDF or DOCX.")

# Extract skills and rank resume
skills = extract_skills(resume_text)
processed_job_desc = process_job_description(job_description)
score = train_model([resume_text], processed_job_desc)[0]

# Display results
print("Extracted Skills:", skills)
print("Resume Score:", round(score, 2))
```

```python
import spacy
import pdfplumber
import docx
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from flask import Flask, request, jsonify
from flask_ngrok import run_with_ngrok
import gradio as gr

# Load NLP model
nlp = spacy.load("en_core_web_sm")

# Flask App
app = Flask(__name__)
run_with_ngrok(app)

# Extract text from PDF
def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = " ".join([page.extract_text() for page in pdf.pages if page.extract_text()])
    return text

# Extract text from DOCX
def extract_text_from_docx(docx_path):
    doc = docx.Document(docx_path)
    return " ".join([para.text for para in doc.paragraphs])

# Extract Skills using NLP
def extract_skills(text):
    doc = nlp(text)
    return list(set(token.text.lower() for token in doc if token.pos_ in ["NOUN", "PROPN"]))

# Process job description
def process_job_description(job_desc):
    doc = nlp(job_desc)
    return " ".join([token.lemma_ for token in doc if not token.is_stop])

# Train TF-IDF model
def train_model(resumes, job_desc):
    vectorizer = TfidfVectorizer()
    job_vec = vectorizer.fit_transform([job_desc])
    resume_vecs = vectorizer.transform(resumes)
    scores = cosine_similarity(resume_vecs, job_vec).flatten()
    return scores

# API Route for Resume Processing
@app.route('/process_resume', methods=['POST'])
def process_resume():
    file = request.files['resume']
    job_desc = request.form['job_description']

    # Extract text
    if file.filename.endswith('.pdf'):
        text = extract_text_from_pdf(file)
    elif file.filename.endswith('.docx'):
        text = extract_text_from_docx(file)
    else:
        return jsonify({'error': 'Invalid file format'}), 400

    # Extract features
    skills = extract_skills(text)
    processed_job_desc = process_job_description(job_desc)

    # Ranking score
    score = train_model([text], processed_job_desc)[0]

    return jsonify({'skills': skills, 'score': round(score, 2)})

# Gradio UI Function
def gradio_interface(file, job_desc):
    if file.name.endswith('.pdf'):
        text = extract_text_from_pdf(file.name)
    elif file.name.endswith('.docx'):
        text = extract_text_from_docx(file.name)
    else:
        return "Invalid file format", ""

    skills = extract_skills(text)
    processed_job_desc = process_job_description(job_desc)
    score = train_model([text], processed_job_desc)[0]

    return f"Skills Extracted: {', '.join(skills)}", f"Resume Score: {round(score, 2)}"

# Launch Gradio UI
demo = gr.Interface(
    fn=gradio_interface,
    inputs=[gr.File(label="Upload Resume (PDF/DOCX)"), gr.Textbox(label="Job Description")],
    outputs=[gr.Textbox(label="Extracted Skills"), gr.Textbox(label="Resume Score")]
)

demo.launch(share=True)

if __name__ == '__main__':
    app.run()
```
